```python
# Print the closing prices of the Google stock  for each day
All_Stocks_Data=df[['Close']].values
#print(All_Stocks_Data[0:5])
print(All_Stocks_Data[0:60])
```

```python
# Feature Scaling for fast training of neural networks


from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
#The MinMaxScaler scales the features to a specific range, usually between 0 and 1.
sc=MinMaxScaler()
```

```python
#Applying feature scaling on dataframe using scaler
Data_Scaler = sc.fit(All_Stocks_Data)
X=Data_Scaler.transform(All_Stocks_Data)
#X=FullData
```

```python
print('** Stock Data  After Normalization **')
#X[0:5]
X[0:60]
```

```python
#Empty list to store input features
X_input_samples = list()

#Empty list to store samples of target variables
y_target_samples = list()
 
```

```python
# Assign the length of the input dataset X  to the variable Num_Rows
Num_Rows = len(X)

#Num_TimeSteps=10  

# Predicting next day's Price using past day's prices
#The number of past day's prices that will be used as input features for predicting the next day's price.
Num_TimeSteps=60
```

```python
# Iterate thru the values 
import pandas as pd
for i in range(Num_TimeSteps, len(X)):
    x_sample = X[i-Num_TimeSteps:i]
    y_sample = X[i]
    X_input_samples.append(x_sample.tolist())
    y_target_samples.append(y_sample.tolist())
```

```python
# Reshape the Input as a 3D (number of samples, Time Steps, Features)
X_data = np.array(X_input_samples)
X_data = X_data.reshape(X_data.shape[0],X_data.shape[1], 1)
print('\n*** Input Data shape ***')
print(X_data.shape)
```

```python
# no need to  reshape y   as it is supposed to be a single column only
y_data=np.array(y_target_samples)
y_data=y_data.reshape(y_data.shape[0], 1)
print('\n#### Output Data shape ####')
print(y_data.shape)
```

```python
# the number of test data records
#Test_Records=5
#Test_Records=60
Test_Records=225   
 
# Splitting the dataset  into train and test set
X_train_data=X_data[:-Test_Records]
X_test_data=X_data[-Test_Records:]
y_train_data=y_data[:-Test_Records]
y_test_data=y_data[-Test_Records:]
 
# Printing the shape of training and testing dataset
print('\n***Training Dataset shape ***')
print(X_train_data.shape)
print(y_train_data.shape)
print('\n*** Testing Dataset shape ***')
print(X_test_data.shape)
print(y_test_data.shape)
```

```python
#visualizing the input-output that are used to train an LSTM 

for input_values, output_values in zip(X_train_data[0:2], y_train_data[0:2]):
    print(input_values,'--', output_values)
```

```python
# Defining Input shapes for LSTM
Num_TimeSteps=X_train_data.shape[1]
Total_Features=X_train_data.shape[2]
print("Number of TimeSteps:", Num_TimeSteps)
print("Number of Features:", Total_Features)
```

```python
# Importing the Keras libraries and packages


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
import time

# Initialising the model
model = Sequential()
 
# Adding the First input hidden layer and the LSTM layer

model.add(LSTM(units = 60, activation = 'relu', input_shape = (Num_TimeSteps, Total_Features), return_sequences=True))

 
# Adding the  Second hidden layer and the LSTM layer
model.add(LSTM(units = 60, activation = 'relu', input_shape = (Num_TimeSteps, Total_Features), return_sequences=True))

 
# Adding the Fourth  hidden layer and the LSTM layer
model.add(LSTM(units = 60, activation = 'relu', return_sequences=False ))

 
model.summary()
  
# Adding the output layer
model.add(Dense(units = 1))

# Compiling the model
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
 

# Measure the time taken by the model for training
Start_Time=time.time()
 
# Fit the model to the Training set
model.fit(X_train_data, y_train_data, batch_size = 32, epochs = 100)
 
End_Time=time.time()
print("## Total Time Taken: ", round((End_Time-Start_Time)/60), 'Minutes ##')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# defining function for the predicted stock price using the model
def generate_predicted_stock_price(last_10_days):
    # Normalizing and reshaping data
    last_10_days_normalized = Data_Scaler.transform(last_10_days.reshape(-1, 1))
    last_10_days_reshaped = last_10_days_normalized.reshape(1, len(last_10_days), 1)

    # Generte the predicted stock price
    predicted_stock_price_normalized = model.predict(last_10_days_reshaped)
    predicted_stock_price = Data_Scaler.inverse_transform(predicted_stock_price_normalized)

    return predicted_stock_price[0][0]


# defining Function to plot the stock prices
def plot_stock_prices(last_10_days, predicted_stock_price):
    
    # Create an array for the x-axis (time steps)
    time_steps_actual = np.arange(1, len(last_10_days) + 1)
    time_steps_predicted = len(last_10_days) + 1

    # Plotting the actual stock prices and the predicted stock price
    plt.plot(time_steps_actual, last_10_days, label='Actual')
    plt.axhline(y=predicted_stock_price, color='red', linestyle='--', label='Predicted')

    # give labels and title to the plot
    plt.xlabel('Time Steps')
    plt.ylabel(' Stock Closing Price')
    plt.title('Actual vs. Predicted Stock Prices')

    # Display the legend
    plt.legend()

    # Display the plot
    plt.show()



def main():
    # data
    last_10_days = df["Close"].values[-10:]

    # Generate the predicted stock price
    predicted_stock_price = generate_predicted_stock_price(last_10_days)

    # Plot the stock prices
    plot_stock_prices(last_10_days, predicted_stock_price)


if __name__ == "__main__":
    main()




```
